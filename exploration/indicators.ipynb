```python
import sys
sys.path.append('../')
from plotting import CandlePlot
import pandas as pd
from technicals.indicators import BollingerBands, ATR, KeltnerChannels, RSI, MACD
```

```python
df = pd.read_pickle("../data/AUD_CAD_H1.pkl")
```

# Bollinger Bands

```python
df_an = df.copy()
```

```python
df_an = BollingerBands(df_an)
```

```python
cp = CandlePlot(df_an.iloc[-200:], candles=True)
cp.show_plot(line_traces=['BB_MA', 'BB_UP', 'BB_LW'])
```

# Average True Range

```python
df_an_2 = df.copy()
df_an_2 = ATR(df_an_2)
```

```python
df_an_2.tail()
```

```python
df_plot = df_an_2.iloc[-200:]
cp = CandlePlot(df_plot, candles=True)
cp.show_plot(sec_traces=['ATR'])
```

# Keltner Channels

```python
df_an_3 = df.copy()
```

```python
df_an_3 = KeltnerChannels(df_an_3)
```

```python
dfp = df_an_3.iloc[-200:]
cp = CandlePlot(dfp, candles=True)
cp.show_plot(line_traces=['EMA', 'KeUp', 'KeLo'])
```

# Relative Strength Index

```python
df_an_4 = df.copy()
df_an_4 = RSI(df_an_4)
```
